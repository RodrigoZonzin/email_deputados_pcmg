```python
import pandas as pd
import re
```

```python
data = pd.read_csv('almg1.csv')
```

```python
y = data[data.columns[-3:]]
```

```python
c = y['Contatos']
```

```python
email_addresses = []

for string in c:
    match = re.search(r'E-mail([^\s]+)', string)
    if match:
        email_addresses.append(match.group(1))

print(email_addresses)
```

```python
y['Emails'] = email_addresses
```

```python
deputados = y['Deputado']
```

```python
siglas = ['PT', 'REDE', 'PMN', 'PL', 'PSD', 'PP', 'PV', 'PDT', 'REPUBLICANOS',
         'UNIÃO', 'PSB', 'PCdoB' 'AVANTE', 'MDB', 'PSOL', 'PSC', 'NOVO', 'PATRIOTA', 'CIDADANIA']
```

```python
deputados
```

```python
pattern = '|'.join(siglas)

# Split the strings based on the pattern
res = [re.split(f'({pattern})', s)[0] for s in deputados]
```

```python
print(res)
```

```python
print(len(res))
```

```python
y['nome_deputado'] = res
```

```python
dado_final = pd.DataFrame(zip(y['Emails'], y['nome_deputado']))
dado_final.columns = ['email', 'nome']
```
